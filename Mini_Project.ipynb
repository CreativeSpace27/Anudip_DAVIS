```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
categories = []
expenses = []

n = int(input("Enter number of expense categories: "))

for i in range(n):
    category = input(f"Enter category {i+1}: ")
    expense = float(input(f"Enter expense for {category}: "))

    categories.append(category)
    expenses.append(expense)

df = pd.DataFrame({
    "Category": categories,
    "Expense": expenses
})

df
```

```python
total_expense = df["Expense"].sum()
average_expense = df["Expense"].mean()
median_expense = df["Expense"].median()

print("Total Expense:", total_expense)
print("Average Expense:", average_expense)
print("Median Expense:", median_expense)
```

```python
high_expenses = df[df["Expense"] > average_expense]
high_expenses
```

```python
plt.figure()
plt.plot(df["Category"], df["Expense"])
plt.xlabel("Category")
plt.ylabel("Expense Amount")
plt.title("Expense Trend by Category")
plt.show()
```

```python
plt.figure()
plt.bar(df["Category"], df["Expense"])
plt.xlabel("Category")
plt.ylabel("Expense Amount")
plt.title("Expense Distribution by Category")
plt.show()
```

```python
budget_limit = 4000

df["Status"] = df["Expense"].apply(
    lambda x: "Over Budget" if x > budget_limit else "Within Budget"
)

df
```

```python
print("------ Expense Summary ------")
print("Total Expense:", total_expense)
print("Average Expense:", average_expense)
print("Median Expense:", median_expense)

print("\nHigh Expense Categories:")
print(high_expenses)
```
